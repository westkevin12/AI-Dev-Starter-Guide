# The Ultimate AI Development Environment Guide

Welcome. This guide is designed to take you from a standard Windows 11 PC to a professional-grade AI Workstation. We assume **zero prior knowledge**. Follow these steps exactly in order.

**Goal:** Safely install Ubuntu Linux on a dedicated drive and configure it with Google's Antigravity IDE, Local AI models, and Professional Cloud tools.

## Part 1: Windows Preparation (Do this FIRST)

Before we touch Linux, we need to prepare your Windows system and create the installation media.

### Step 1.1: Gather Hardware
*   **1x USB Drive (8GB or larger)**: This will be wiped.
*   **1x Dedicated SSD (NVMe or SATA)**: This is where we will install Ubuntu.

### Step 1.2: Download Software
1.  **Download Ubuntu:** [Ubuntu Desktop Download Page](https://ubuntu.com/download/desktop) (Get the LTS ISO).
2.  **Download Rufus:** [Rufus.ie](https://rufus.ie/) (Get the portable version).

### Step 1.3: Create Bootable USB
1.  Plug in USB. Open Rufus.
2.  **Device:** Select USB.
3.  **Boot selection:** Select Ubuntu ISO.
4.  Click **START** (Say Yes/OK to prompts).

### Step 1.4: Critical Windows Settings
1.  **Disable Fast Startup** (Control Panel > Hardware and Sound > Power Options > Choose what buttons do > Uncheck "Turn on fast startup").
2.  **Disable BitLocker** (Optional but recommended for setup).

## Part 2: The Installation (Safety First)

We will use the "Physical Isolation" method.

### Step 2.1: Disconnect Windows
**Shut down** and **unplug the Windows Drive**. This guarantees you cannot accidentally wipe Windows.

### Step 2.2: Boot from USB
Plug in Ubuntu USB -> Power On -> Press Boot Menu Key (F12/F8/F11/F9) -> Select USB -> "Try or Install Ubuntu".

### Step 2.3: Install Ubuntu
1.  **Install Ubuntu** -> English -> English (US).
2.  **Updates:** Normal installation + Install third-party software.
3.  **Installation Type:** Select **"Erase disk and install Ubuntu"** (Safe because Windows is unplugged).
4.  Finish installation -> Restart.

### Step 2.4: Reconnect Windows
Shut down -> Reconnect Windows Drive -> Power On -> Set Ubuntu as first boot option in BIOS.

### Step 3.4: Install GitHub CLI
The official tool to login to GitHub without messing with SSH keys.

```python
# 1. Install
sudo mkdir -p -m 755 /etc/apt/keyrings \
&& wget -qO- https://cli.github.com/packages/githubcli-archive-keyring.gpg | sudo tee /etc/apt/keyrings/githubcli-archive-keyring.gpg > /dev/null \
&& chmod go+r /etc/apt/keyrings/githubcli-archive-keyring.gpg \
&& echo "deb [arch=$(dpkg --print-architecture) signed-by=/etc/apt/keyrings/githubcli-archive-keyring.gpg] https://cli.github.com/packages stable main" | sudo tee /etc/apt/sources.list.d/github-cli.list > /dev/null \
&& sudo apt update \
&& sudo apt install gh -y

# 2. Login
# Note: This is interactive. Run it in a terminal if it hangs here.
gh auth login
```

### Step 4.2: Install Pyenv (Python Version Manager)
Control your Python version per project. Never break your system Python again.

```python
# 1. Install Pyenv from official script
curl https://pyenv.run | bash

# 2. Add to your shell (Copy ALL lines at once)
echo 'export PYENV_ROOT="$HOME/.pyenv"' >> ~/.bashrc
echo '[[ -d $PYENV_ROOT/bin ]] && export PATH="$PYENV_ROOT/bin:$PATH"' >> ~/.bashrc
echo 'eval "$(pyenv init -)"' >> ~/.bashrc

# 3. Apply changes
source ~/.bashrc

# 4. Install Python 3.11
# Note: You might need to restart your terminal/notebook kernel for pyenv to be found
export PYENV_ROOT="$HOME/.pyenv"
export PATH="$PYENV_ROOT/bin:$PATH"
eval "$(pyenv init -)"

pyenv install 3.11.9
pyenv global 3.11.9

# 5. Verify
python --version
```

### Step 4.3: Install NVM (Node Version Manager)
Avoid "permission denied" errors and switch Node versions easily.

```python
# 1. Install NVM
curl -o- https://raw.githubusercontent.com/nvm-sh/nvm/v0.39.7/install.sh | bash

# 2. Add to shell
export NVM_DIR="$HOME/.nvm"
[ -s "$NVM_DIR/nvm.sh" ] && \. "$NVM_DIR/nvm.sh"

# 3. Install latest LTS node
nvm install --lts

# 4. Verify
node -v
```

### Step 4.5: Install Google Antigravity IDE
The official AI-first IDE by Google.

```python
# 1. Setup Keyrings
sudo mkdir -p /etc/apt/keyrings
curl -fsSL https://us-central1-apt.pkg.dev/doc/repo-signing-key.gpg | sudo gpg --dearmor --yes -o /etc/apt/keyrings/antigravity-repo-key.gpg

# 2. Add Repository
echo "deb [signed-by=/etc/apt/keyrings/antigravity-repo-key.gpg] https://us-central1-apt.pkg.dev/projects/antigravity-auto-updater-dev/ antigravity-debian main" | sudo tee /etc/apt/sources.list.d/antigravity.list > /dev/null

# 3. Install
sudo apt update
sudo apt install antigravity
```

### Step 4.6: Install Docker Engine
Docker is essential for running vector databases and containerized AI agents.

```python
# 1. Add Docker's official GPG key:
sudo apt-get update
sudo apt-get install -y ca-certificates curl
sudo install -m 0755 -d /etc/apt/keyrings
sudo curl -fsSL https://download.docker.com/linux/ubuntu/gpg -o /etc/apt/keyrings/docker.asc
sudo chmod a+r /etc/apt/keyrings/docker.asc

# 2. Add the repository to Apt sources:
echo \
  "deb [arch=$(dpkg --print-architecture) signed-by=/etc/apt/keyrings/docker.asc] https://download.docker.com/linux/ubuntu \
  $(. /etc/os-release && echo "$VERSION_CODENAME") stable" | \
  sudo tee /etc/apt/sources.list.d/docker.list > /dev/null
sudo apt-get update

# 3. Install Docker
sudo apt-get install -y docker-ce docker-ce-cli containerd.io docker-buildx-plugin docker-compose-plugin

# 4. Critical: Run Docker without sudo
sudo usermod -aG docker $USER
newgrp docker
```
